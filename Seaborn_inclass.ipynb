<b>Visualizing Statistical Relationships:</b>
The process of understanding relationships between variables of a dataset and how these relationships,
<br>in turn, depend on other variables is known as statistical analysis. 


relplot():
<br>This is a figure-level-function that makes use of two other axes functions for Visualizing Statistical Relationships which are:

<br>scatterplot()
<br>lineplot()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### Scatter plot

```python
f = sns.load_dataset("flights")

sns.relplot(x="passengers", y="month", data=f)
```

### lineplot()

```python
a=pd.DataFrame({'Day':[1,2,3,4,5,6,7],'Grocery':[30,80,45,23,51,46,76],'Clothes':[13,40,34,23,54,67,98],'Utensils':[12,32,27,56,87,54,34]},index=[1,2,3,4,5,6,7])
g = sns.relplot(x="Day", y="Clothes", kind="line", data=a)

a
```

### Plotting with Categorical Data:
catplot():with different Kind parameters

```python

sns.set(style="ticks", color_codes=True)
a = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=a);
a
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
a = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", kind="violin", data=a)
```

### Multiple Grid Plots


```python
sns.set(style="darkgrid")
a = sns.load_dataset("iris")
b = sns.FacetGrid(a, col="species")
b.map(plt.hist, "sepal_length");
```

```python
sns.set(style="white", color_codes=True)
a = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=a);
```

```python
#### style ::darkgrid, ticks, whitegrid.
sns.set(style="darkgrid", color_codes=True)
a = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=a);
```
